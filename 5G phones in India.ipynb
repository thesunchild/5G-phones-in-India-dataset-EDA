```python
df= df.drop( ['price in first site', 'price in second site'], axis=1) 

df.head()
```

```python
df.info()
```

```python
df.rename(columns={'real price available': 'price'}, inplace= True)
df.head()
```

```python
def commadelete(value):
    value = str(value)
    if ',' in value:
        value=value.replace(',' , '')
        return float(value)
    else:
        return float(value)
    
df['price']=df['price'].apply(commadelete)
df['price'].unique()
```
